```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Function to generate the waveform and spectrogram
def plot_waveform_and_spectrogram(audio_files, labels):
    fig, axs = plt.subplots(2, 3, figsize=(15, 8))

    for i, file in enumerate(audio_files):
        # Load the audio file
        y, sr = librosa.load(file, sr=None)

        # Plot the waveform (top row)
        axs[0, i].plot(y)
        axs[0, i].set_title(f'{labels[i]} - Waveform')
        axs[0, i].set_xlabel('Time')
        axs[0, i].set_ylabel('Amplitude')

        # Plot the spectrogram (bottom row)
        D = librosa.amplitude_to_db(librosa.stft(y), ref=np.max)
        librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log', ax=axs[1, i])
        axs[1, i].set_title(f'{labels[i]} - Spectrogram')
        axs[1, i].set_xlabel('Time')
        axs[1, i].set_ylabel('Frequency')

    plt.tight_layout()
    plt.show()

# Define your audio file paths (replace these with actual paths)
audio_files = [
    'F:\\Awais_data\\Datasets\\asvspoof2019\\LA\\ASVspoof2019_LA_train\\flac\\LA_T_1138215.flac',  # Path to real audio
    'F:\\Awais_data\\Datasets\\asvspoof2019\\LA\\ASVspoof2019_LA_train\\flac\\LA_T_8495510.flac',  # Path to fake audio
    'F:\\Awais_data\\Datasets\\PartialSpoof\\Train\\con_wav\\CON_T_0015777.wav'  # Path to partial fake audio
]

# Define the corresponding labels
labels = ['Real', 'Fake', 'Partial Fake']

# Generate the waveform and spectrogram plots
plot_waveform_and_spectrogram(audio_files, labels)
```
